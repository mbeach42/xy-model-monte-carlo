```julia
include("xyModelMC.jl");
using PyPlot
```

```julia
L = 8
N_eq = 10
N = 100
T_low = 0.1
T_high = 3.0
num_Ts = 30
T_range = linspace(T_low, T_high, num_Ts)
```

```julia
clusters = 10
dir_name = "C:\\Users\\allgu\\Dropbox (Personal)\\wolff-cluster-julia\\xy-data\\"
@time E, M, Cv, X, Stiff = xyModelMC.iterate_over_temperatures(T_low, T_high, num_Ts, L, N_eq, N, clusters, "wolff", dir_name);
```

```julia
clusters = 10
dir_name = "none"
@time E_metro, M_metro, Cv_metro, X_metro, Stiff_metro = xyModelMC.iterate_over_temperatures(T_low, T_high, num_Ts, L, N_eq, N, clusters, "metro", dir_name);
```

```julia
xlabel("Temperature")
ylabel("Energy")
plot(T_range, E, ".")
plot(T_range, E_metro, ".")
```

```julia
xlabel("Temperature")
ylabel("Magnetization")
plot(linspace(T_low, T_high, num_Ts), M, ".")
plot(linspace(T_low, T_high, num_Ts), M_metro, ".")
```

```julia
xlabel("Temperature")
ylabel("Cv")
plot(linspace(T_low, T_high, num_Ts), Cv, ".")
plot(linspace(T_low, T_high, num_Ts), Cv_metro, ".")
```

```julia
xlabel("Temperature")
ylabel("X")
plot(linspace(T_low, T_high, num_Ts), X, ".")
plot(linspace(T_low, T_high, num_Ts), X_metro, ".")
```

```julia
xlabel("Temperature")
ylabel("Stiffness")
plot(linspace(T_low, T_high, num_Ts), Stiff, ".")
plot(linspace(T_low, T_high, num_Ts), Stiff_metro, ".")
```
